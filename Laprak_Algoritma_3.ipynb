```python
print("Program pembelanjaan USAKTIF")
nama = input("Masukkan nama:")
nim = input("Masukkan nim: ")
total_belanja = int(input("Masukkan total belanja anda: Rp "))
pakai_kupon = input("Masukkan kode kupon? y/n : ")

potongan = 0
diskon = 0

if pakai_kupon.lower() == "y":
    kode = input("Masukkan kode kupon:")
    if kode == "IKL6309":
      print("---Selamat!Anda mendapatkan diskon 100% (gratis)---")
      potongan = total_belanja
      total_bayar = 0
    else:
        print("Kode kupon salah/expired")
        total_bayar = total_belanja

else:
    if total_belanja >= 390000:
      diskon = 80
    elif total_belanja >= 190000:
      diskon = 60
    elif total_belanja >= 90000:
      diskon = 40
    else:
      diskon = 0

    potongan = total_belanja * (diskon / 100)
    total_bayar = total_belanja - potongan

    if diskon == 100:
      print("---SELAMAT ANDA MENDAPATKAN DISKON 100%---")

print(f"Total yang harus dibayar: Rp {total_bayar}")
print("---TERIMAKASIH SUDAH BERBELANJA DI USAKTIF---")
```
