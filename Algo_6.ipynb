```python
"""
#Annisa Qurrotu
#065002500031
#Latihan1
"""
def konversi(huruf='A,B,C,D'):
    if huruf == 'A':
        return 4
    elif huruf == 'B':
        return 3
    elif huruf == 'C':
        return 2
    elif huruf == 'D':
        return 1
    else:
        return 0

# fungsi menghitung rata-rata
def rata_rata(n1, n2):
    return (n1 + n2) / 2

# program utama
h1 = input("Masukkan nilai: ").upper()
n1 = konversi(h1)
print("Nilai =", n1)

h2 = input("Masukkan nilai: ").upper()
n2 = konversi(h2)
print("Nilai =", n2)

print("Rata-ratanya adalah:", rata_rata(n1, n2))
```

```python
"""
#Annisa Qurrotu
#065002500031
#Latihan2
"""
def kabisat(tahun):
    return (tahun % 4 == 0 and tahun % 100 != 0) or (tahun % 400 == 0)

def jumlah_hari(bulan, tahun):
    if bulan == 2:
        if kabisat(tahun):
            return 29
        else:
            return 28
    elif bulan in [4, 6, 9, 11]:
        return 30
    else:
        return 31

print("Program ini akan menentukan jumlah hari dalam bulan tertentu")
print("Masukkan 0 untuk menghentikan program")

while True:
    bulan = int(input("masukkan bulan (1-12): "))
    if bulan == 0:
        print("Terima kasih sudah menggunakan program saya, sampai berjumpa lagi.")
        break
    tahun = int(input("masukkan tahun (e.g., 2021): "))
    print(f"{jumlah_hari(bulan, tahun)} hari dalam sebulan\n")
```
