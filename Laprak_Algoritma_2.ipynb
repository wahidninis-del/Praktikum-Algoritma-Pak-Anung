```python
"""
*Annisa Qurrotu Aini Wahid
*065002500031
"""
import math
print("Silahkan isi lattitude dan longitudenya untuk menentukan jarak antara 2 kota yang dicari:")
R = 6371
kota1 = input("Masukkan kota1 :")
kota2 = input("Masukkan kota2 :")

lat1 = float(input("Masukkan lat1:"))
lon1 = float(input("Masukkan lon1:"))
lat2 = float(input("Masukkan lat2:"))
lon2 = float(input("Masukkan lon2:"))

lat1_rad = math.radians(lat1)
lon1_rad = math.radians(lon1)
lat2_rad = math.radians(lat2)
lon2_rad = math.radians(lon2)

dlat = lat2_rad - lat1_rad
dlon = lon2_rad - lon1_rad

a = math.sin(dlat/2)**2 + math.cos(lat1_rad) * math.cos(lat2_rad) * math.sin(dlon/2)**2
c = 2 * math.asin(math.sqrt(a))

jarak = R * c

print(f"Jarak antara {kota1} dan {kota2} adalah {jarak} kilometer")
```
